```python
# Dependencies and Setup
import pandas as pd

# CSV Files to Load 
school_data= "Resources/schools_complete.csv"
student_data= "Resources/students_complete.csv"

# Reading School and Student CSV Data Files and storing them into Pandas DataFrames
school_data_df = pd.read_csv(school_data)
student_data_df = pd.read_csv(student_data)

# Combining the data into a single dataset.  
school_data_complete = pd.merge(student_data_df, school_data_df, how="left", on=["school_name", "school_name"])
school_data_complete.head()
```

```python
#Calculating the total number of unique schools
schoolcount = len(school_data_complete["school_name"].unique())
schoolcount
```

```python
#Calculating the total number of students
studentcount = len(school_data_complete["student_name"])
studentcount
```

```python
#Calculating the total budget
totalbudget = school_data_complete["budget"].unique().sum()
totalbudget
 
```

```python
#Calculating the average math score and reading score
averagemathscore = float(school_data_complete["math_score"].sum())/float(len(school_data_complete["math_score"]))
averagereadingscore = float(school_data_complete["reading_score"].sum())/float(len(school_data_complete["reading_score"]))
print(averagemathscore)
print(averagereadingscore)
```

```python
# Calculating the percentage of students who passed math (math scores greather than or equal to 70)
passingmathcount = school_data_complete[(school_data_complete["math_score"] >= 70)].count()["student_name"]
passingmathpercentage = passingmathcount / float(studentcount) * 100
passingmathpercentage
```

```python
# Calculating the percentage of students who passed reading scores greater than equal to 70
passingreadingcount = school_data_complete[(school_data_complete["reading_score"] >= 70)].count()["student_name"]
passingreadingpercentage = passingreadingcount / float(studentcount) * 100
passingreadingpercentage
```

```python
#Calculating the percentage of students that passed math and reading
passingmathandreadingcount = school_data_complete[
    (school_data_complete["math_score"] >= 70) & (school_data_complete["reading_score"] >= 70)
].count()["student_name"]
overallpassingrate = passingmathandreadingcount /  float(studentcount) * 100
overallpassingrate
```

```python
# Creating a high-level snapshot summary of the district's key metrics in a DataFrame
districtsummary = {"Total Schools": schoolcount,
"Total Students":studentcount,
"Total Budget": totalbudget,
"Average Math Score": averagemathscore,
"Average Reading Score": averagereadingscore,
"% Passing Math":passingmathpercentage,
"% Passing Reading":passingreadingpercentage,
"% Overall Passing":overallpassingrate}
district_summary_df = pd.DataFrame(districtsummary,index=[0])
# Formatting total students and total budget to show numbers with comma separators eg: '10,000' instead of '10000'
district_summary_df["Total Students"] = district_summary_df["Total Students"].map("{:,}".format)
district_summary_df["Total Budget"] = district_summary_df["Total Budget"].map("${:,.2f}".format)

# Displaying the DataFrame of the district data summary
district_summary_df.head()
```

```python
#Setting School Names as the index for out data set
schooltypes = school_data_df.set_index(["school_name"])['type']
```

```python
#Calculating total student count
per_school_counts = school_data_complete.groupby('school_name').size()
print(per_school_counts)
```

```python
# Calculating the total school budget and per capita spending
per_school_budget = school_data_complete.groupby(["school_name"]).mean()["budget"]
per_school_capita = per_school_budget / per_school_counts
```

```python
#Calculating Average test scores per school
per_school_math = school_data_complete.groupby(["school_name"]).mean()["math_score"]
per_school_reading = school_data_complete.groupby(["school_name"]).mean()["reading_score"]
```

```python
#Calculating the percentage of students with math scores of 70 or higher per school
school_passing_math = (school_data_complete[school_data_complete['math_score']>=70].groupby('school_name')["Student ID"].count()/per_school_counts*100)
```

```python
#Calculating the percentage of students with reading scores of 70 or higher per school
school_passing_reading = (school_data_complete[school_data_complete['reading_score']>=70].groupby('school_name')["Student ID"].count()/per_school_counts*100)
```

```python
#Calculating percentage of students who passed math and reading per school
per_school_passing_math_and_reading = school_data_complete[(school_data_complete["reading_score"] >= 70) & (school_data_complete["math_score"] >= 70)].groupby('school_name')["Student ID"].count()/per_school_counts*100
```

```python
# Creating a DataFrame called `per_school_summary` with columns for the calculations above.
perschoolsummary = {"School Type": schooltypes,
"Total Students":per_school_counts,
"Total School Budget":per_school_budget,
"Per Student Budget":per_school_capita,
"Average Math Score":per_school_math,
"Average Reading Score":per_school_reading,
"% Passing Math":per_school_math,
"% Passing Reading":per_school_reading,
"% Overall Passing":per_school_passing_math_and_reading}
per_school_summary = pd.DataFrame(perschoolsummary)
# Formatting numbers to have comma notation
per_school_summary["Total School Budget"] = per_school_summary["Total School Budget"].map("${:,.2f}".format)
per_school_summary["Per Student Budget"] = per_school_summary["Per Student Budget"].map("${:,.2f}".format)

# Displaying the DataFrame of School Data displaying a summary of key metrics for each school
per_school_summary
```

```python
#Creating a dataframe of the top 5 performing schools based on % overall passing
topschools = per_school_summary.sort_values("% Overall Passing",ascending=False)
topschools.head()
```

```python
#Creating a dataframe of the bottom 5 performing schools based on % Overall Passing
lowestschools = per_school_summary.sort_values("% Overall Passing")
lowestschools.head()
```

```python
#Sorting Original Data By Grades
ninth_graders = school_data_complete[(school_data_complete["grade"] == "9th")]
tenth_graders = school_data_complete[(school_data_complete["grade"] == "10th")]
eleventh_graders = school_data_complete[(school_data_complete["grade"] == "11th")]
twelfth_graders = school_data_complete[(school_data_complete["grade"] == "12th")]

#Taking the mean of each grade per school
ninth_graders_math_scores = school_data_complete[(school_data_complete["grade"] == "9th")].groupby('school_name').mean()['math_score']
tenth_graders_math_scores = school_data_complete[(school_data_complete["grade"] == "10th")].groupby('school_name').mean()['math_score']
eleventh_graders_math_scores = school_data_complete[(school_data_complete["grade"] == "11th")].groupby('school_name').mean()['math_score']
twelfth_graders_math_scores = school_data_complete[(school_data_complete["grade"] == "12th")].groupby('school_name').mean()['math_score']


math_scores_by_grade = pd.DataFrame({"9th": ninth_graders_math_scores,
"10th":tenth_graders_math_scores,
"11th":eleventh_graders_math_scores,
"12th":twelfth_graders_math_scores})

math_scores_by_grade.index.name = None

math_scores_by_grade
```

```python
#Taking the average reading scores per grade
ninth_graders_reading_scores = school_data_complete[(school_data_complete["grade"] == "9th")].groupby('school_name').mean()['reading_score']
tenth_graders_reading_scores = school_data_complete[(school_data_complete["grade"] == "10th")].groupby('school_name').mean()['reading_score']
eleventh_graders_reading_scores = school_data_complete[(school_data_complete["grade"] == "11th")].groupby('school_name').mean()['reading_score']
twelfth_graders_reading_scores = school_data_complete[(school_data_complete["grade"] == "12th")].groupby('school_name').mean()['reading_score']


# Combining each of the average scores above into single DataFrame called `reading_scores_by_grade`
reading_scores_by_grade = pd.DataFrame({"9th":ninth_graders_reading_scores,
'10th':tenth_graders_reading_scores,
'11th':eleventh_graders_reading_scores,
'12th':twelfth_graders_reading_scores})

# Setting the index column name to blank
reading_scores_by_grade.index.name = None

# Displaying the DataFrame
reading_scores_by_grade
```

```python
#Establishing groupings to break down school performance based on spending range
spending_bins = [0,590,620,645,680]
groups = ["<$590","$590-$620","$620-$645","$645-$680"]

#Copying the dataframe of per school summary data to create a new dataframe that shows how spending affects scores
#Output we want to display in the new dataframe is Average Math Score	Average Reading Score	% Passing Math	% Passing Reading	% Overall Passing, with school type as the index

school_spending_df = pd.DataFrame(perschoolsummary)
school_spending_df["Spending Ranges (Per Student)"] = pd.cut(school_spending_df["Per Student Budget"],spending_bins,labels=groups)
school_spending_grouped = school_spending_df.groupby("Spending Ranges (Per Student)").mean()

#Deleting columns of data we don't want to output on the new dataframe table
del school_spending_grouped["Total Students"]
del school_spending_grouped["Total School Budget"]
del school_spending_grouped["Per Student Budget"]
school_spending_grouped
```

```python
#Setting up groupings to break down scores by school size
size_bins = [0, 1000, 2000, 5000]
labels = ["Small (<1000)", "Medium (1000-2000)", "Large (2000-5000)"]

#Copying the per school dataframe to group the data into a new dataframe that shows how school size affects scores
#Output we want to display in the new dataframe is Average Math Score	Average Reading Score	% Passing Math	% Passing Reading	% Overall Passing, with school type as the index

school_size_df = per_school_summary.copy()
school_size_df["School Size"] = pd.cut(school_size_df['Total Students'],size_bins,labels=labels)
school_size_grouped = school_size_df.groupby("School Size").mean()

#Deleting columns of data we don't want to output on the new dataframe table
del school_size_grouped["Total Students"]

#Displaying new dataframe showing scores by school size
school_size_grouped
```

```python
#Grouping the data in the per school dataframe by school type and taking the mean (average) of the data in it
#Output we want to display is the Average Math Score ,Average Reading Score,	% Passing Math,	% Passing Reading,	% Overall Passing, with school type as the index

grouped_school_type = per_school_summary.copy()
grouped_school_type = grouped_school_type.groupby("School Type").mean()

#Deleting the unwanted column of data from the new dataframe
del grouped_school_type["Total Students"]

grouped_school_type
```
